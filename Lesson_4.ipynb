##9c. "df": Customizing Subplots
---

```python
#run this

df.head(10)
```

**Customizing Figures**

First, create a Subplot with 4 figures = [0][0] to [1][1]

* [0][0] = 1st Figure
>* Plot only the 'col2'
>* Changing the line color to purple
>* Changing the Font Size of the Axis

* [0][1] = 2nd Figure
>* For "df", the index column is from a - j
>* if we `enumerate(df['col2'].index`, we would obtain...
>>* a: 0
>>* b: 1
>>* c: 2
>>* ...
 
>* Presume we only want to show a / c / e / g / i (the even numbered)
>* `idx % 2 == 0` will help us only obtain the even numbered

>* We plot only the the rows a/c/e/g/i

* [1][0] = 3rd Figure
>* Plot all columns
>* Put the Axis Font Size = 20
>* Rotate the Axis Font by -90 degrees

* [1][1] = 4th Figure
>* Plot all columns
>* Leave the Axis Font Size = 20
>* Put the Title Font Size = 21
>* Rotate the Axis Font by -50 degrees
>* We use `.set_title()` to change title size


```python
#run this

#Create a Subplot
fig, axes = plt.subplots(2,2, figsize=(16,8))

# [0][0]
df['col2'].plot(figsize=(16,4), color='purple', fontsize=21, ax=axes[0][0])

# [0][1]
ticks_to_show = [idx for idx, _ in enumerate(df['col2'].index) if idx % 2 == 0] #even id

df['col2'].plot(figsize=(16,4), color='purple', xticks=ticks_to_show, fontsize=16, ax=axes[0][1])

# [1][0]
df.plot(figsize=(15,7), title='Tiny Title',
        fontsize=20, rot=-90, ax=axes[1][0])\

# [1][1] 
df.plot(figsize=(16,8), fontsize=20, rot=-50, ax=axes[1][1])\
        .set_title('Better-Sized Title', fontsize=21, y=1.01)
```
